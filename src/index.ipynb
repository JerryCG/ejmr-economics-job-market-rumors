# Examples

Below are some examples that can be used from the EJMR dataset. If you see the website, each post has title. From that, we can generate what are the most common words used as the title. From that, we may be able to see what's being discussed among those anonymous economists. Below I present a bar chart which consists of 5 most common words as the title. Please note as the notebook is automatically updated hence the values of the top five are not fixed.

```python
word = ejmr.tokenize_word()
ejmr.bar_chart(ejmr.top_five_word(word), 'count', 'words', '5 most common words in EJMR title', 'word', 'example-1', True)
```

Next, we can find the distributions of posts and views for each thread. The numbers are in log-scale. Vertical lines show median values.

```python
ejmr.distributions('example-2', True)
```

Not only that, we can see the relationship between numbers of posts and views per thread. However, this does not imply any causality.

```python
ejmr.relationship('example-3', True)
```

Lastly, we may want to see how many posts are there on EJMR per year? Since it provides posting time, though not exactly (e.g. 1 year ago, 3 months ago, 5 years ago, etc), we are able to plot a time series like the following.

```python
ejmr.post_per_year('example-4', True)
```
